```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Input data points
x = np.arange(-10, 10, 0.1)
```

```python
# Plotting function
def plot_graph(x, y, title='Activation Function'):
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel(title+'(x)')
    plt.grid(True)
    plt.show()
```

```python
# Sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))

y = sigmoid(x)
plot_graph(x, y, 'Sigmoid')
```

```python
# Hyperbolic Tangent
def tanh(x):
    return (2 * sigmoid(2*x)) - 1

y = tanh(x)
plot_graph(x, y, 'Hyperbolic Tangent')
```

```python
# Softmax
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))

y = softmax(x)
print('Sum of output values from Softmax function: ',np.sum(y))
plot_graph(x, y, 'Softmax')
```

```python
# Softsign
def softsign(x):
    return (x/(np.abs(x)+1))

y = softsign(x)
plot_graph(x, y, 'Softsign')
```

```python
# ReLU
def re_lu(x):
    result = []
    for i in x:
        if i<0:
            i = 0
        result.append(i)
    return result

y = re_lu(x)
plot_graph(x, y, 'ReLU')
```

```python
# Leaky ReLU
def leaky_relu(x):
    result = []
    for i in x:
        if i<0:
            i = 0.01*i
        result.append(i)
    return result

y = leaky_relu(x)
plot_graph(x, y, 'Leaky ReLU')
```

```python
# Parametric ReLU
def param_relu(x, a=0.1):
    result = []
    for i in x:
        if i<0:
            i = a*i
        result.append(i)
    return result

y = param_relu(x, a=0.1)
plot_graph(x, y, 'Parametric ReLU')
```

```python
# Softplus
def softsign(x):
    return np.log(1+np.exp(x))

y = softsign(x)
plot_graph(x, y, 'Softplus')
```

```python
# Exponential Linear Unit
def exp_lu(x, a=1):
    result = []
    for i in x:
        if i<0:
            i = a*(np.exp(i)-1)
        result.append(i)
    return result

y = exp_lu(x, a=1)
plot_graph(x, y, 'Exponential Linear Unit')
```

```python
# Scaled Exponential Linear Unit
def s_elu(x, a=1.6733, t=1.0507):
    result = []
    for i in x:
        if i<0:
            i = t*(a*(np.exp(i)-1))
        result.append(i)
    return result

y = s_elu(x, a=1)
plot_graph(x, y, 'Scaled Exponential Linear Unit')
```

```python
# Swish Function
def swish(x):
    return x*(1/(1+np.exp(-x)))

y = swish(x)
plot_graph(x, y, 'Swish')
```

```python
import tensorflow as tf # Using tensorflow math function

# Mish Function
def mish(x):
    return x * tf.math.tanh(tf.math.softplus(x))

y = mish(x)
plot_graph(x, y, 'Mish')
```
