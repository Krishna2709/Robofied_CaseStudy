```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Input data points
x = np.arange(-10, 10, 0.1)
```

```python
# Plotting function
def plot_graph(x, y, title='Activation Function'):
    plt.figure(dpi=150)
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel(title+'(x)')
    plt.grid(True)
    plt.show()
```

### 1. Identity Function

```python
# Define the Identity function
def identity(x):
    return x

y = identity(x)
plot_graph(x, y, 'Identity Function')
```

### 2. Sigmoid Function

```python
# Define the Sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

y = sigmoid(x)
plot_graph(x, y, 'Sigmoid')
```

### 3. Hyperbolic Tangent Function

```python
# Define the Hyperbolic Tangent function
def tanh(x):
    return (2 * sigmoid(2*x)) - 1

y = tanh(x)
plot_graph(x, y, 'Hyperbolic Tangent')
```

### 4. Softmax Function

```python
# Define the Softmax function
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))

y = softmax(x)
print('Sum of output values from Softmax function: ',np.sum(y))
plot_graph(x, y, 'Softmax')
```

### 5. Softsign Function

```python
# Define the Softsign function
def softsign(x):
    return (x/(np.abs(x)+1))

y = softsign(x)
plot_graph(x, y, 'Softsign')
```

### 6. ReLU function

```python
# Define the ReLU function
def re_lu(x):
    result = []
    for i in x:
        if i<0:
            i = 0
        result.append(i)
    return result

y = re_lu(x)
plot_graph(x, y, 'ReLU')
```

### 7. Leaky ReLU Function

```python
# Define the Leaky ReLU function
def leaky_relu(x):
    result = []
    for i in x:
        if i<0:
            i = 0.01*i
        result.append(i)
    return result

y = leaky_relu(x)
plot_graph(x, y, 'Leaky ReLU')
```

### 8. Parametric ReLU Function

```python
# Define the Parametric ReLU function
def param_relu(x, a=0.1):
    result = []
    for i in x:
        if i<0:
            i = a*i
        result.append(i)
    return result

y = param_relu(x, a=0.1)
plot_graph(x, y, 'Parametric ReLU')
```

### 9. Softplus Function

```python
# Define the Softplus function
def softsign(x):
    return np.log(1+np.exp(x))

y = softsign(x)
plot_graph(x, y, 'Softplus')
```

### 10. Exponential Linear Unit Function 

```python
# Define the Exponential Linear Unit function
def exp_lu(x, a=1):
    result = []
    for i in x:
        if i<0:
            i = a*(np.exp(i)-1)
        result.append(i)
    return result

y = exp_lu(x, a=1)
plot_graph(x, y, 'Exponential Linear Unit')
```

### 11. Scaled Exponential Linear Unit Function

```python
# Define the Scaled Exponential Linear Unit function
def s_elu(x, a=1.6733, t=1.0507):
    result = []
    for i in x:
        if i<0:
            i = t*(a*(np.exp(i)-1))
        result.append(i)
    return result

y = s_elu(x, a=1)
plot_graph(x, y, 'Scaled Exponential Linear Unit')
```

### 12. Swish Function

```python
# Define the Swish function 
def swish(x):
    return x*(1/(1+np.exp(-x)))

y = swish(x)
plot_graph(x, y, 'Swish')
```

### 13. Mish Function

```python
import tensorflow as tf # Using tensorflow math function

# Define the Mish function
def mish(x):
    return x * tf.math.tanh(tf.math.softplus(x))

y = mish(x)
plot_graph(x, y, 'Mish')
```

### Comparing Mish and Swish functions

```python
import tensorflow as tf

# Input data
x = np.arange(-10, 10, 0.001)

# Mish function
m = x * tf.math.tanh(tf.math.softplus(x))
# Swish function
sw = x*(1/(1+np.exp(-x)))

plt.figure(dpi=150)
plt.plot(x,m, label='Mish', color='blue', linestyle='dashed')
plt.plot(x,sw, label='Swish', color='orange')
plt.title('Mish and Swish Comparision')
plt.legend()
plt.show()
```
